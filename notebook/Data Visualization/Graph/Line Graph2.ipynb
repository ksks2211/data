```python
import pandas as pd


df = pd.read_csv("../../../csv/tfr-in-singpore-by-groups.csv",  index_col='Year')
```

```python
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
plt.rc('font', size=14)

for ethnic_group in df.columns:
    plt.plot(df.index, df[ethnic_group], label=ethnic_group, linewidth=3)
    plt.legend(fontsize=14)


plt.grid(True)
plt.ylim(0,3)
plt.xlim(df.index.min(), (df.index.max()//5+1)*5)
plt.title('TFR of Singpore by ethnic groups')
plt.xlabel('Year', fontsize=16)
plt.ylabel('TFR', fontsize=16)

plt.gca().set_facecolor('#f9f9f9') 


plt.savefig('./tfr-in-singpore-by-ethnic-groups.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df.index, df['Malays']- df['Chinese'], label="Malays & Chinese diff", linewidth=3)
plt.legend(fontsize=14)

plt.grid(True)
plt.ylim(0,3)
plt.xlim(df.index.min(), (df.index.max()//5+1)*5)
plt.title('Malays & Chinese diff')

plt.show()
```

```python
df = pd.read_csv("../../../csv/tfr-in-singpore-by-groups.csv",  index_col='Year')
df.drop('Total', axis=1, inplace=True)



plt.figure(figsize=(13,7))
plt.rc('font', size=14)


markers = ['o', 's', '^']
for ethnic_group,marker in zip(df.columns, markers):
    plt.plot(df.index, df[ethnic_group], label=ethnic_group, linewidth=3, marker=marker, markersize=8)
    plt.legend(fontsize=14)



plt.ylim(0.6,3)
plt.xlim(2013, (df.index.max()+1))
plt.title('TFR of Singpore by ethnic groups(2010 ~ )')
plt.xlabel('Year', fontsize=16)
plt.ylabel('TFR', fontsize=16)

plt.gca().set_xticks(range(2013, df.index.max() + 1, 1))

ax = plt.gca()

ax.spines['top'].set_visible(False)  #   
ax.spines['right'].set_visible(False)  #   
ax.spines['bottom'].set_linewidth(2)  #    
ax.spines['left'].set_linewidth(2)  #    

plt.show()
```

```python
from matplotlib import font_manager

font_properties = font_manager.FontProperties(fname="./Roboto-Medium.ttf", size=12)

df_sing = pd.read_csv("../../../csv/tfr-in-singpore-by-groups.csv",  index_col='Year')
df_sing.drop('Total', axis=1, inplace=True)


df_mal = pd.read_csv("../../../csv/tfr-in-malaysia-by-groups.csv",  index_col='Year')
df_mal.drop('Total', axis=1, inplace=True)


plt.figure(figsize=(13,7))


markers = ['o', 's', '^']

colors = ['#d62728',  '#2ca02c', '#ff7f0e']
linestyles = ['-', '--', '-.', ':'] 
for ethnic_group,marker, color in zip(df_sing.columns, markers, colors):
    plt.plot(df_sing.index, df_sing[ethnic_group], label=ethnic_group+"(Singpore)", linewidth=2, marker=marker, markersize=8, color=color, linestyle=':')
    plt.legend(fontsize=14, prop=font_properties)

markers = ['o', 's', '^']
for ethnic_group,marker, color in zip(df_mal.columns, markers, colors):
    plt.plot(df_mal.index, df_mal[ethnic_group], label=ethnic_group+"(Malaysia)", linewidth=2, marker=marker, markersize=8, color=color)
    plt.legend(fontsize=14, prop=font_properties)


plt.ylim(0.6,3.3)
plt.xlim(2013, (df.index.max()+1))
plt.title('TFR of Malaysia & Singapore by ethnic groups(2010 ~ )', fontproperties=font_properties, fontsize=18)
plt.xlabel('Year', fontsize=16, fontproperties=font_properties)
plt.ylabel('TFR', fontsize=16, fontproperties=font_properties)

plt.gca().set_xticks(range(2013, df.index.max() + 1, 1))


ax.legend(fontsize=12, prop=font_properties, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)

ax = plt.gca()

ax.spines['top'].set_visible(False)  #   
ax.spines['right'].set_visible(False)  #   
ax.spines['bottom'].set_linewidth(2)  #    
ax.spines['left'].set_linewidth(2)  #    

plt.show()
```
