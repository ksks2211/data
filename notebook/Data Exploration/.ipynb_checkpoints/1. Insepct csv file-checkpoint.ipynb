```python
# Correlation matrix
print(df.corr())

# Visualize the correlation matrix
import seaborn as sns

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Pairwise scatter plots
sns.pairplot(df)
plt.show()
```

```python
df['population_moving_average'] = df['population'].rolling(window=5).mean()

# Plot the moving average
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['population_moving_average'], marker='o')
plt.title('5-Year Moving Average of Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
# Calculate year-on-year percentage change for population
df['population_pct_change'] = df['population'].pct_change() * 100

# Plot the percentage change
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['population_pct_change'], marker='o')
plt.title('Year-on-Year Percentage Change in Population')
plt.xlabel('Year')
plt.ylabel('Percentage Change')
plt.grid(True)
plt.show()
```

```python
# Find the year with the lowest TFR
year_lowest_tfr = df.loc[df['tfr'].idxmin(), 'year']
print(f"The year with the lowest TFR is: {year_lowest_tfr}")
```

```python
# Create a sub-dataframe where TFR is below 2.0
df[df['tfr'] < 2.0]
```

```python
# Create a condition where births exceed deaths by a certain number

df['growth'] = df['births'] - df['deaths']


condition = df['growth'] > 1000000  # Example threshold

# Apply condition to create a sub-dataframe
significant_birth_excess = df[condition]

significant_birth_excess= significant_birth_excess.sort_values(by = 'growth', ascending=False)
significant_birth_excess[['year','growth', 'births', 'deaths']]
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def millions_formatter(x, pos):
    return f'{x / 1_000_000}M'

def thousands_formatter(x, pos):
    return f'{x / 1_000}K'

def billions_formatter(x, pos):
    return f'{x / 1_000_000_000}B'

plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['population'], marker='o', linestyle='-')
plt.title('Population Over Time')
plt.xlabel('Year')
plt.ylabel('Population')

# Set the y-axis formatter
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(millions_formatter))

plt.show()
```

```python
us_tfr_by_years =df['tfr']
us_tfr_by_years.index = df['year']
```

```python
us_tfr_by_years
```

```python
import numpy as np
plt.figure(figsize=(9, 5.5))
plt.plot(
    us_tfr_by_years, 
    color='green', 
    linewidth=3,
    linestyle='-'
)
plt.title(f'United States({df['year'].min()} ~ {df['year'].max()})', fontweight='bold')

# X-axis
plt.xlim(df['year'].min()-2,df['year'].max()+4)
plt.xlabel("Year", size=12)
plt.xticks(np.arange(df['year'].min(),df['year'].max()+3,10))

# Y-axis
plt.ylabel("TFR",size=12)

h = df['tfr'].max() - df['tfr'].min()
plt.ylim(df['tfr'].min()-h*0.1,df['tfr'].max() + h*0.1)


plt.grid(True)
plt.legend(labels=['TFR'],loc='best',fontsize=12)


plt.show()
```
